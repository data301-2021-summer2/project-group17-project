```python
# importing csv dataset

import pandas as pd
cetacean = pd.read_csv("../../data/raw/allCetaceanData.csv")
cetacean
```

```python
# drop the column mother
cetacean = cetacean.drop('mother',axis='columns')
```

```python
# drop the column father
cetacean = cetacean.drop('father',axis='columns')
```

```python
# drop the column region, all US, redundant
cetacean = cetacean.drop('region',axis='columns')
```

```python
cetacean
```

```python
# removing column transfer
cetacean = cetacean.drop('transfer',axis='columns')
```

```python
# removing column transferDate
cetacean.drop('transferDate',axis='columns')
```

```python
# removing the notes column
cetacean = cetacean.drop('notes', axis = 'columns')
```
